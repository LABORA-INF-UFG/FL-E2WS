```python
# Convert images to one-dimensional arrays
num_train_samples = train_images.shape[0]
num_test_samples = test_images.shape[0]
num_features = train_images.shape[1] * train_images.shape[2]

train_images_reshaped = train_images.reshape((num_train_samples, num_features))
test_images_reshaped = test_images.reshape((num_test_samples, num_features))

# Create DataFrames for training and testing data
df_train = pd.DataFrame(train_images_reshaped)
df_test = pd.DataFrame(test_images_reshaped)

# Add a column for labels at the end
df_train['label'] = train_labels
df_test['label'] = test_labels
```

```python
def create_model():
    model = tf.keras.models.Sequential(
        [
            tf.keras.layers.Dense(units=128, activation='relu', input_shape=(784,)), 
            tf.keras.layers.Dense(10, activation="softmax")
        ]
    )

    model.compile(
        optimizer=tf.keras.optimizers.Adam(),
        loss=tf.keras.losses.SparseCategoricalCrossentropy(),
        metrics=['accuracy']
    )

    return model
```

```python
def create_model():
    model = tf.keras.models.Sequential([

        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(28, 28, 1)), 
        tf.keras.layers.MaxPooling2D((2, 2), strides=(2, 2)),
        tf.keras.layers.Conv2D(16, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.MaxPooling2D((2, 2), strides=(2, 2)),
        
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
    ])
    
    model.compile(optimizer=tf.keras.optimizers.Adam(),
            loss=tf.keras.losses.SparseCategoricalCrossentropy(),
            metrics=['accuracy']
    )
    
    return model
    
x_train = np.array([x.reshape(28,28) for x in x_train.reset_index(drop=True).values])
x_test = np.array([x.reshape(28,28) for x in x_test.reset_index(drop=True).values])
```

```python
path_clients="/home/renan/Documentos/FL/datasets/fashion-mnist/non-iid-0.9-100-rotation-45"
```

```python
n = 100
lista = [[] for _ in range(n)]

for i in range(n):
    model = create_model()
    
    
    df_train = pd.read_pickle(f"{path_clients}/{i+1}_train.pickle")
    df_test = pd.read_pickle(f"{path_clients}/{i+1}_test.pickle")
    
    x_train = df_train.drop(['label'], axis=1)
    y_train = np.array(df_train['label'])
    x_train = np.array([x.reshape((28, 28, 1)) for x in x_train.reset_index(drop=True).values])
               
    x_test = df_test.drop(['label'], axis=1)
    x_test = np.array([x.reshape((28, 28, 1)) for x in x_test.reset_index(drop=True).values])
    y_test = np.array(df_test['label'])

   
    
    model.fit(x_train, y_train,
                batch_size=128,
                epochs=30,
                validation_data=(x_test, y_test),
                verbose=0)
    print(f"{i}")
    for j in range(n):
        df_test_evaluate = pd.read_pickle(f"{path_clients}/{j+1}_test.pickle")
        x_test_evaluate = df_test_evaluate.drop(['label'], axis=1)
        x_test_evaluate = np.array([x.reshape((28, 28, 1)) for x in x_test_evaluate.reset_index(drop=True).values])
        y_test_evaluate = np.array(df_test_evaluate['label'])        
        loss, accuracy = model.evaluate(x_test_evaluate, y_test_evaluate, verbose=0)
        lista[i].append(accuracy)

print(lista)
```

```python
import seaborn as sns
import pandas as pd
df = pd.DataFrame(lista)
sns.set(rc = {'figure.figsize':(5,4)})
b = sns.heatmap(df, cmap="rocket", annot=False, fmt='g', annot_kws = {'size':12}, vmin=0, vmax=1)

b.set_xlabel("NIID R-FMNIST",fontsize=16)
b.set_ylabel("Accuracy of $\it{w_{i}}$ for \n each $P_{i}$ of N",fontsize=16)
b.tick_params(labelsize=14)

cbar = b.collections[0].colorbar
cbar.ax.tick_params(labelsize=14)
```
